```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
ruta = "Data/clima.parquet"

clima = pd.read_parquet(ruta)
```

Uso la columna "" para referenciar el viaje y le agrego las 3 columnas de datos referida al clima

```python
clima.columns
```

```python
filtro_columnas = ['tpep_dropoff_datetime', 'date', 'temperature_2m', 'precipitation_mm']

clima = clima[filtro_columnas]
```

```python
clima
```

```python
# Crear el boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=clima[['temperature_2m', 'precipitation_mm']])
plt.title('Boxplot de Temperatura y Precipitación')
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.show()
```

temperatura_2m, significa que son datos registrados a 2m en la ciudad de NY

Asumo que la temp es en Grados Centigrados, por los valores

Temperatura normal (promedio): La temperatura normal promedio en la ciudad de Nueva York varía según la estación. En invierno, la temperatura normal puede estar alrededor de los 0 °C a 5 °C (32 °F a 41 °F), mientras que en verano puede oscilar entre los 20 °C a 25 °C (68 °F a 77 °F).

Temperatura máxima: Durante los meses de verano, la temperatura máxima promedio puede rondar los 30 °C a 35 °C (86 °F a 95 °F). Sin embargo, en días extremadamente calurosos, la temperatura máxima puede superar los 38 °C (100 °F) en algunas ocasiones.

Temperatura mínima: Durante los meses de invierno, la temperatura mínima promedio puede descender a alrededor de -5 °C a 0 °C (23 °F a 32 °F). En noches extremadamente frías, las temperaturas mínimas pueden caer por debajo de -10 °C (14 °F).

```python
# Crear el histograma de temperatura
plt.figure(figsize=(10, 6))
plt.hist(clima['temperature_2m'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de Temperatura en NYC')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
