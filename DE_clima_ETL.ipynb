```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
ruta = "Data/clima.parquet"

clima = pd.read_parquet(ruta)
```

Uso la columna "" para referenciar el viaje y le agrego las 3 columnas de datos referida al clima

```python
clima.columns
```

```python
filtro_columnas = ['tpep_dropoff_datetime', 'date', 'temperature_2m', 'precipitation_mm']

clima = clima[filtro_columnas]
```

```python
clima
```

```python
# Crear el boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=clima[['temperature_2m', 'precipitation_mm']])
plt.title('Boxplot de Temperatura y PrecipitaciÃ³n')
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.show()
```

temperatura_2m, significa que son datos registrados a 2m en la ciudad de NY

Asumo que la temp es en Grados Centigrados, por los valores

Temperatura normal (promedio): La temperatura normal promedio en la ciudad de Nueva York varÃ­a segÃºn la estaciÃ³n. En invierno, la temperatura normal puede estar alrededor de los 0 Â°C a 5 Â°C (32 Â°F a 41 Â°F), mientras que en verano puede oscilar entre los 20 Â°C a 25 Â°C (68 Â°F a 77 Â°F).

Temperatura mÃ¡xima: Durante los meses de verano, la temperatura mÃ¡xima promedio puede rondar los 30 Â°C a 35 Â°C (86 Â°F a 95 Â°F). Sin embargo, en dÃ­as extremadamente calurosos, la temperatura mÃ¡xima puede superar los 38 Â°C (100 Â°F) en algunas ocasiones.
